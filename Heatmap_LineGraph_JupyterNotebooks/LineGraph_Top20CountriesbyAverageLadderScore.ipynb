## Brief Summary

This line graph shows the changes in Ladder scores over time for the top 20 countries. The graph indicates that the Ladder scores have been consistently high for certain countries such as Finland, Denmark, and Switzerland, while others such as Greece and Italy have shown more variation. Overall, there seems to be a slight upward trend in Ladder scores for most countries in the top 20, with some countries showing more significant increases over time.

```python
# import the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a pandas dataframe
df = pd.read_csv("C:\\Users\\cryst\\Documents\\UMN_Bootcamp\\Coursework\\Section_2_Data_Analytics_Python\\Module_7\\group_5_project_1\\CSV_Files\\DataForTable2.1WHR2023.csv")
```

```python
# create a new dataframe with the top 20 countries based on their ranking
top_20 = df.groupby(['Country name']).agg({'Life Ladder': 'mean'}).reset_index().sort_values('Life Ladder', ascending=False).head(20)
```

```python
# pivot the dataframe to create a multi-line graph
pivoted = df.pivot(index='year', columns='Country name', values='Life Ladder')
pivoted = pivoted[top_20['Country name']] # filter only the top 20 countries
```

```python
# plot the graph
pivoted.plot(figsize=(12,8))
plt.title('Top 20 Countries by Average Ladder Score')
plt.xlabel('Year')
plt.ylabel('Ladder Score')
plt.savefig('line_graph.png', dpi=300, bbox_inches='tight')
plt.show()
```
