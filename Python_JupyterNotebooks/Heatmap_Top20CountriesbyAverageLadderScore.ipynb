## Brief Summary

This heatmap shows the relationship between the Top 20 countries and their Ladder scores over time. The countries are sorted by their average life ladder score, with Finland having the highest average score. The heatmap reveals that most of the top 20 countries have had an increase in their life ladder score over time, with a few exceptions. It also shows that the life ladder scores for most countries tend to be higher in the more recent years, which suggests an overall improvement in subjective well-being over time.

```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file into a pandas dataframe
df = pd.read_csv("C:\\Users\\cryst\\Documents\\UMN_Bootcamp\\Coursework\\Section_2_Data_Analytics_Python\\Module_7\\group_5_project_1\\CSV_Files\\DataForTable2.1WHR2023.csv")
```

```python
# Get the top 20 countries by average life ladder score
top_20 = df.groupby('Country name')['Life Ladder'].mean().sort_values(ascending=False).head(20).index.tolist()
```

```python
# Filter the data to include only the top 20 countries
filtered_data = df[df['Country name'].isin(top_20)]
```

```python
# Pivot the data to have years as columns and countries as rows
pivoted_data = filtered_data.pivot(index='Country name', columns='year', values='Life Ladder')
```

```python
# Create a heatmap using seaborn
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(pivoted_data, cmap='YlGnBu')
plt.title('Top 20 Countries by Average Ladder Score')
plt.xlabel('Year')
plt.ylabel('Country')
heatmap.figure.savefig("heatmap.png", dpi=300)
plt.show()
```
